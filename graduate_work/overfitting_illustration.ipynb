```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [i*np.pi/2 for i in np.arange(-5,5,)]
```

```python
sinx_x = [i for i in np.arange(-8,8, 0.01)]
```

```python
sin5x_x = [5*i for i in np.arange(-8,8, 0.01)]
```

```python

plt.scatter(data, np.sin(data))
plt.plot(sinx_x, np.sin(sinx_x), c = 'b', label = 'sin(x)')
plt.plot(sinx_x, np.sin(sin5x_x), c = 'r', label = 'sin(5x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
